```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Part 1
mpg = sns.load_dataset('mpg')
sns.heatmap(mpg.corr(), cmap='coolwarm')
sns.pairplot(mpg)
```

```python
#Part 2
dia = sns.load_dataset('diamonds')
dia = dia[dia.color != 'D']
dia = dia[dia.color != 'E']
dia = dia[dia.cut != 'Fair']
    
g2 = sns.FacetGrid(dia, col='cut', row='color')
g2 = g2.map(plt.scatter, 'price', 'carat' )
```

```python
#Part 3
crash = sns.load_dataset('car_crashes')
g3 = sns.regplot(x=crash['speeding'], y=crash['total'])
```

```python
g4 = sns.regplot(x=crash['alcohol'], y=crash['total'])
```

```python
#Part 4
iris = sns.load_dataset('iris')

fig, axes = plt.subplots(2, 2)

ax1 = sns.boxplot('species', 'petal_length', data = iris, ax=axes[0,0])
ax2 = sns.boxplot('species', 'petal_width', data = iris, ax=axes[0,1])
ax3 = sns.boxplot('species', 'sepal_length', data = iris, ax=axes[1,0])
ax4 = sns.boxplot('species', 'sepal_width', data = iris, ax=axes[1,1])
```
