```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
labels = ['H', 'He', 'Li', 'Be', 'B', 'C']
atomicWeights = [1,4,7,9,11,12]
```

Part 1:

```python
explode = (0,0,0,0,.1,0)

plt.pie(atomicWeights, labels = labels, explode =explode, autopct = lambda x: '{:.0f}'.format(x * sum(atomicWeights) / 100.0))
;
```

```python
explode = (0,0,0,0,0,.1)

plt.pie(atomicWeights, labels = labels, explode=explode, autopct = '%1.1f%%');

```

Part 2:

```python
df = pd.read_csv('py_ide2.csv')
fig, ax = plt.subplots(2, figsize = (14,8))
ax[0].barh(df['IDE'], df['Adoption'])
ax[0].set_title('IDE\'s Popularity')
ax[0].set_xlabel('Percent Adoption')
ax[0].set_ylabel('IDE')

plt.xticks(rotation=45)
ax[1].bar(df['IDE'], df['Adoption'])
ax[1].set_title('IDE\'s Popularity')
ax[1].set_ylabel('Percent Adoption')
ax[1].set_xlabel('IDE')
plt.tight_layout()
```

Part 3:

```python
days = ['2020-01-01', '2020-02-1', '2020-03-01','2020-04-01','2020-05-01','2020-06-01','2020-07-01','2020-08-01']
array = np.random.uniform(100,200,8)

df2 = pd.DataFrame(list(zip(days, array)), columns = ['Days', 'Numbers'])
df2['Days'] = pd.to_datetime(df2['Days'])
df2.set_index('Days', inplace = True)
```

```python
fig, ax = plt.subplots(2, figsize = (10,8))
ax[0].plot(df2.index, df2['Numbers'])
ax[0].set_title('Days V. Numbers')
ax[0].set_xlabel('Days')
ax[0].set_ylabel('Numbers')

ax[1].bar(df2.index, df2['Numbers'], width = 5)
ax[1].set_title('Days V. Numbers')
ax[1].set_xlabel('Days')
ax[1].set_ylabel('Numbers')
plt.tight_layout()
```

```python
df3 = pd.read_csv('TSLA.csv')

fig = plt.figure(figsize = (15,5))

ax1 = fig.add_axes([0.1, .6, .85, .4])
ax2 = fig.add_axes([.1, .25, .85, .15])
plt.xticks(rotation = 75)
ax1.plot(df3['Date'], df3['Close'], 'r', label = 'Close')
ax2.ticklabel_format(style='plain')

ax2.plot(df3['Date'], df3['Volume'], 'g', label = 'Volume')



ax1.set_title('Closing Price of TSLA')
ax2.set_title('Volume of TSLA')
ax1.set_ylabel('Price')
ax2.set_ylabel('Volume')
ax1.set_xlabel('Date')
ax2.set_xlabel('Date')
```
